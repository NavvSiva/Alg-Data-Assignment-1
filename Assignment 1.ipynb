```python
import time
Data = 'product_data.txt'

Array = []

with open(Data, 'r') as file:
    for line in file:
        row = line.strip().split(',')
        product = {
            "ID": row[0],
            "Name": row[1],
            "Price": float(row[2]),
            "Category": row[3]
        }
        Array.append(product)

print("Product Data List:")
for product in Array:
    print(product)



def insert_product(Array, new_product):
    Array.append(new_product)

new_product = {"ID": "00001", "Name": "iPhone 18", "Price": 999.99, "Category": "Electronics"}
insert_product(Array, new_product)
print("\nInserted New Product at the End:")
for product in Array:
    print(product)



def update_product(Array, product_update, product_id):
    for product in Array:
        if product["ID"] == product_id:
            product.update(product_update)
            return True
    return False

update = {"Name": " MacBook", "Price": 3000.99}
update_product(Array, update, "00001")
print("")
print("Updated a Product in the List:")
for product in Array:
    print(product)



def delete_product(Array, product_id):
    for i in range(len(Array)):
        if Array[i]["ID"] == product_id:
            Array.pop(i)
            return True
    return False
        
delete_product(Array, "00001")
print("")
print("A Product has been Deleted:")
for product in Array:
    print(product)



def search(Array, key, value):
    for product in Array:
        if product[key] == value:
            return product
    return None

searched_product = search(Array, "ID","44574")
print("")
print("Results for Searched ID# :")
print(searched_product)

searched_product = search(Array, "Name", " Knife Set ASRHX")
print("")
print("Results for Searched Name:")
print(searched_product)

searched_product = search(Array, "Price", 64.62)
print("")
print("Results for Searched Price: ")
print(searched_product)

searched_product = search(Array, "Category", " Home & Kitchen")
print("")
print("Results for Searched Category: ")
print(searched_product)


def BubbleSort_price(Array):
    n = len(Array)
    for i in range(n):
        for j in range (0, n-i-1):
            if Array[j]["Price"] > Array[j+1]["Price"]:
                Array[j], Array[j+1] = Array[j+1], Array[j]

BubbleSort_price(Array)
print("")
print("Results after Bubble Sort by Price:")
for product in Array:
    print(product)


def ComplexityAnalysis(data, description, time_complexity):
    print(f"\n{description} - Starting array: ")
    start = time.time()
    BubbleSort_price(data.copy())
    end = time.time()
    print(f"{description} - Time taken: {end - start:.6f} seconds")
    print(f"Time Complexity: {time_complexity}\n")

sorted_data = sorted(Array, key=lambda x: x['Price'])
reverse_data = sorted(Array, key=lambda x: x['Price'], reverse=True)


ComplexityAnalysis(sorted_data, "Sorted data", "O(n^2)")
ComplexityAnalysis(reverse_data, "Reverse order data", "O(n^2)")
```
